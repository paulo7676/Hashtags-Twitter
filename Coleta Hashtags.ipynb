```python
#Recebe um soup do dia
def selecionar_tabelas(Table01):
    lista_tweets = []

    for i in Table01:
        #período
        hora = i.find("div").text
        #Tabela em tempo X
        trending = i.find_all("tr",class_="tr_table")
        quantidade = i.find_all("tr",class_="tr_table1")
        
        #Valores da tabela
        for w in range(len(trending)):
            novo_tweet = []
            #adicionando a hora
            novo_tweet.append(hora)

            #adicionando o tweet que entrou no trending atual
            trending_atual = trending[w].text
            novo_tweet.append(trending_atual[trending_atual.index(")")+1:])

            #adicionando a quantidade de tweets naquele período
            lista_tweets.append(novo_tweet)
            novo_tweet.append(quantidade[w].text)
    return lista_tweets

def data_dia(soup):
    b = soup.find("h1")
    try:
        data= b.text
        data2 = data[data.index("(")+1:len(data)-1]
    except AttributeError:
        return 0
    return data2

#seleciona todos os trending_topics_unicos
def all_keyword_from_day(lista_tweets):
    trending_dia = []
    for i in lista_tweets:
        if i[1] not in trending_dia:
            trending_dia.append(i[1])
            
    return trending_dia

def topics_dia_frases(lista_tweets,data,nome_arquivo):
    
    with open(nome_arquivo, 'a', newline='', encoding='UTF-8') as csvfile:
        fieldnames = ['Data','Trending','Quantidade_Tweets','Horario_inicio','Horario_final']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        
        if(data != 0):
            for a in all_keyword_from_day(lista_tweets):
                primeira_ocorrencia = 0
                ultima_ocorrencia = 0
                quantidade_de_tweets = 0
                for b in lista_tweets:
                    if a == b[1]:
                        if primeira_ocorrencia == 0:
                            primeira_ocorrencia = b[0]
                            ultima_ocorrencia = b[0]
                            quantidade_de_tweets = b[2]
                        else:
                            ultima_ocorrencia = b[0]
                            quantidade_de_tweets = b[2]
                writer.writerow({'Data':data,
                                'Trending':a,
                                'Quantidade_Tweets':quantidade_de_tweets,
                                'Horario_inicio':primeira_ocorrencia,
                                'Horario_final':ultima_ocorrencia})
        else:
            print("Erro_Dia")
             
```
